# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Extraction

```python
# Ghi những giá trị cần phân tích vào file csv và đọc
df = pd.read_csv('example.csv')
df.head()
```

```python
df.info()
```

# Data Cleaning & Processing

```python
# Thay đổi title của những người không để tên (trừ các thầy)
# sort lại DataFrames theo Number
df.loc[df.Title == 'ex', 'Title'] = 'Nguyễn Văn Mạnh 22022521'
df.loc[df.Title == 'btvn', 'Title'] = 'Vũ Vân Long 22022501'
df.loc[df.Title == 'Week(6)', 'Title'] = 'Long Trí Thái Sơn 22022653'
df.loc[df.Title == 'Create README.md', 'Title'] = 'Nguyễn Kim Hoàng Anh 22022566'
df = df.sort_values('Number').reset_index(drop = True)
```

# Data Analysis

```python
# Người có số lần commit nhiều nhất
df[df.Commits == df.Commits.max()][['Title', 'Commits']]
```

```python
# Người có số dòng code thêm vào nhiều nhất (đóng góp nhiều nhất)
df[df.Additions == df.Additions.max()][['Title', 'Additions']]
```

```python
# Người có số file thay đổi nhiều nhất
df[df['Number of Files Changed'] == df['Number of Files Changed'].max()][['Title', 'Number of Files Changed']]
```

```python
# Tuần có số lần commit nhiều nhất
dict1 = {}
for date in df['Commit Date']:
    lst = [x.strip(" '") for x in date.strip("[]").split(",")]
    for item in lst:
        if item in dict1:
            dict1[item] += 1
        else:
            dict1[item] = 1
weekCommit = list(filter(lambda x : dict1[x] == max(dict1.values()), dict1.keys()))
weekCommit
```

```python
# Những từ xuất hiện nhiều nhất trong commit messages
dict2 = {}
for messages in df['Commit Message']:
    word_list = messages.strip('][').split(', ')
    for word in word_list:
        list1 = word.split()
        for w in list1:
            if w in dict2:
                dict2[w] += 1
            else:
                dict2[w] = 1
mostWord = list(filter(lambda x : dict2[x] == max(dict2.values()), dict2.keys()))
mostWord
```

```python
# File bị thay đổi nhiều nhất
dict3 = {}
for files in df['Files Changed']:
    lst = [x.strip(" '[].") for x in files.strip("[]").split(",")]
    for file in lst:
        if file in dict3:
            dict3[file] += 1
        else:
            dict3[file] = 1
mostFileChanged = list(filter(lambda x : dict3[x] == max(dict3.values()), dict3.keys()))
mostFileChanged
```

# Data Visualization

```python
fig, ax = plt.subplots(figsize = (14, 6))
data = pd.melt(df[['Title', 'Commits', 'Number of Files Changed']], id_vars=['Title'], var_name='variable', value_name='value')
sns.barplot(x='Title', y='value', hue='variable', data = data)
ax.set_title('số Commit và số File thay đổi của mỗi người')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90);
```

```python
fig, ax = plt.subplots(figsize = (14, 6))
sns.barplot(data = df, x = 'Title', y = 'Additions')
ax.set_title('mức độ đóng góp của mỗi người')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90);
```

```python
fig, ax = plt.subplots(1, 2, figsize = (14, 6))
sns.scatterplot(data = df, x = 'Commits', y = 'Number of Files Changed', ax = ax[0], color = 'red')
ax[0].set_title('Mối liên hệ giữa số lần Commit và số File thay đổi')
sns.scatterplot(data = df, x = 'Commits', y = 'Additions', ax = ax[1])
ax[1].set_title('Mối liên hệ giữa số lần Commit và mức độ đóng góp')
```
